```python
import pandas as pd
df = pd.read_csv("/content/finance-data/Finance_data.csv")
print(df.head())
```

```python
print(df.columns.tolist())
print("\nColumn info:")
print(df.dtypes)
```

```python
# Check null values with percentages
null_check = df.isnull().sum()
null_percentage = (df.isnull().sum() / len(df)) * 100

# Create a summary DataFrame
null_summary = pd.DataFrame({
    'Null_Count': null_check,
    'Null_Percentage': null_percentage
})

print("NULL VALUES SUMMARY:")
print(null_summary)
```

```python
# Check for duplicate rows
duplicate_rows = df.duplicated()
total_duplicates = duplicate_rows.sum()

print(f"Total duplicate rows: {total_duplicates}")
print(f"Percentage of duplicates: {(total_duplicates / len(df) * 100):.2f}%")
```

```python
print(df.shape)
```

```python
# Get all object columns
object_columns = df.select_dtypes(include=['object']).columns

print("UNIQUE VALUES IN OBJECT COLUMNS:")
print("=" * 50)

for col in object_columns:
    unique_values = df[col].unique()
    print(f"\n{col}:")
    print(f"  Total unique values: {len(unique_values)}")
    print(f"  Values: {list(unique_values)}")
```

```python


# Create a function to categorize age ranges
def categorize_age(age):
    if age <= 25:
        return 'Youngers'
    elif age <= 55:
        return 'Mids'
    else:
        return 'Adults'

# Add the new age_range column
df['age_range'] = df['age'].apply(categorize_age)

# Display the updated dataframe
print("DataFrame with Age Range Column:")
print(df[['gender', 'age', 'age_range', 'Investment_Avenues']].head(10))

# Show the distribution of age ranges
print("\nAge Range Distribution:")
print(df['age_range'].value_counts().sort_index())

# Show investment status by age range
print("\nInvestment Status by Age Range:")
print(pd.crosstab(df['age_range'], df['Investment_Avenues']))

# Show the complete dataframe info
print(f"\nDataset Shape: {df.shape}")
print(f"Total records: {len(df)}")
```
